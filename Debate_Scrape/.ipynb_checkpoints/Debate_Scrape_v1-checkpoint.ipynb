```python
import numpy as np
import nltk
```

## Part 0: Import Speech Data

```python
debate = open("Obama_McCain_1.txt", encoding="utf8").read().split()
```

## Part 1: Separation Of Moderator, And The Two Candidates

```python
mod_list = []
can1_list = []
can2_list = []

i_new = -1
i_old = -1
while i_new >= -len(debate):
    if debate[i_new] == 'LEHRER:':
        mod_list.append(debate[i_new:])
        del debate[i_new:i_old]
        i_old = i_new

    elif debate[i_new] == 'MCCAIN:':
        can1_list.append(debate[i_new:])
        del debate[i_new:i_old]
        i_old = i_new

    elif debate[i_new] == 'OBAMA:':
        can2_list.append(debate[i_new:])
        del debate[i_new:i_old]
        i_old = i_new

    else: pass
    i_new -= 1
```
